````python
df_2020_USA_50k_50mi = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
````

#remove (USA) from event name

````python
df_2020_USA_50k_50mi['Event name'].str.split('(').str.get(0)
````

````python
df_2020_USA_50k_50mi['Event name'] = df_2020_USA_50k_50mi['Event name'].str.split('(').str.get(0)
````

````python
df_2020_USA_50k_50mi.head(10)
````

````python
# Clean up athlete age
````

````python
df_2020_USA_50k_50mi['athlete_age'] = 2020 - df_2020_USA_50k_50mi['Athlete year of birth']
````

## remove h from athlete performance

````python
df_2020_USA_50k_50mi['Athlete performance'] = df_2020_USA_50k_50mi['Athlete performance'].str.split(' ').str.get(0)
````

````python
df_2020_USA_50k_50mi.head(5)
````

### Drop Columns: Athlete Club, Athlete Country, Athlete year of birth, Athlete Age Category

````python
df_2020_USA_50k_50mi = df_2020_USA_50k_50mi.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
````

````python
df_2020_USA_50k_50mi.head(10)
````

````python
# Clean up null values
````

````python
df_2020_USA_50k_50mi.isna().sum()
````

````python
df_2020_USA_50k_50mi[df_2020_USA_50k_50mi['athlete_age'].isna()==1]
````

````python
df_2020_USA_50k_50mi = df_2020_USA_50k_50mi.dropna()
````

````python
df_2020_USA_50k_50mi.shape
````

### Checking for duplicate values

````python
df_2020_USA_50k_50mi[df_2020_USA_50k_50mi.duplicated() == True]
````

# no dupes

````python
# reset index
````

````python
df_2020_USA_50k_50mi.reset_index(drop = True)
````

````python
#fix types
````

````python
df_2020_USA_50k_50mi.dtypes
````

````python
df_2020_USA_50k_50mi['athlete_age'] = df_2020_USA_50k_50mi['athlete_age'].astype(int)
````

````python
df_2020_USA_50k_50mi['Athlete average speed'] = df_2020_USA_50k_50mi['Athlete average speed'].astype(float)
````

````python
df_2020_USA_50k_50mi.dtypes
````

````python
df_2020_USA_50k_50mi.head(10)
````

````python
#rename columns
````

````python
#Year of event                  int64
##Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#athlete_age                    int64
````

````python
df_2020_USA_50k_50mi = df_2020_USA_50k_50mi.rename(columns = { 'Year of event' : 'year',
                                                              'Event dates' : 'race_day',
                                                              'Event name' : 'race_name',
                                                              'Event distance/length' : 'race_length',
                                                              'Event number of finishers' : 'race_number_of_finishers',
                                                              'Athlete performance' : 'athlete_performance',
                                                              'Athlete gender' : 'athlete_gender',
                                                              'Athlete average speed' : 'athlete_average_speed',
                                                              'Athlete ID' : 'athlete_id'
                                                              })
````

````python
df_2020_USA_50k_50mi.head()
````

````python
#reorder columns
````

````python
df_reordered_2020_USA_50k_50mi = df_2020_USA_50k_50mi[['race_day', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed']]
````

````python
df_reordered_2020_USA_50k_50mi.head()
````

````python
#charts and graphs
````

````python
sns.histplot(df_reordered_2020_USA_50k_50mi['race_length'])
````

````python
sns.histplot(df_reordered_2020_USA_50k_50mi, x = 'race_length', hue = 'athlete_gender')
````

````python
#distribution of average speed
sns.displot(df_reordered_2020_USA_50k_50mi[df_reordered_2020_USA_50k_50mi['race_length'] == '50mi']['athlete_average_speed'])
````

````python
sns.lmplot(data = df_reordered_2020_USA_50k_50mi, x='athlete_age', y='athlete_average_speed', hue='athlete_gender')
````

````python
#Difference in speed for the 50k, 50mi male to female
````

````python
df_reordered_2020_USA_50k_50mi.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
````

````python
#what age groups are the best in the 50m race
````

````python
df_reordered_2020_USA_50k_50mi.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19').head(15)
````

````python
#slowest age groups in the 50m race
````

````python
df_reordered_2020_USA_50k_50mi.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count>19').head(15)
````

````python
#Seasons for the data -> Slower in the summer than winter?

#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2

#split between two decimals
````

````python
df_reordered_2020_USA_50k_50mi['race_month'] = df_reordered_2020_USA_50k_50mi['race_day'].str.split('.').str.get(1).astype(int)
````

````python
df_reordered_2020_USA_50k_50mi['race_season'] = df_reordered_2020_USA_50k_50mi['race_month'].apply(lambda x: "Winter" if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else "Winter")
````

````python
df_reordered_2020_USA_50k_50mi.head()
````

````python
df_reordered_2020_USA_50k_50mi.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
````
